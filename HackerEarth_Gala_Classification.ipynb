```
import os
import shutil

def make_if_dont_exist(folder_path,overwrite=False):
    """
    creates a folder if it doesnot exists
    input: 
    folder_path : relative path of the folder which needs to be created
    over_write :(default: False) if True overwrite the existing folder 
    """
    if os.path.exists(folder_path):
        
        if not overwrite:
            print('Folder already exists.')
        else:
            print(f"{folder_path} overwritten")
            shutil.rmtree(folder_path)
            os.makedirs(folder_path)

    else:
      os.makedirs(folder_path)
      print(f"{folder_path} created!")

folder = './'
folder_path = os.path.join(folder,'train')

for cls in classes:
  make_if_dont_exist(os.path.join(folder_path,cls),overwrite = False)
  
```

# **Normalization Parameters**

If a network in its architecture uses batch normalization, then normalized values of the images needs to be passed while training (e.g. Resnet's family).

What I will be doing is known as finding the normalization parameters over the entire dataset. By first finding the **mean of the whole training set** and then calculating the std. dev. But **batch-wise mean and std.** can also be calculated and a different value than this can be expected in that case.

```
#finding the mean

#dataset loader
dataset = datasets.ImageFolder(('train'),transform = transforms.ToTensor())
mean = 0.0

for img,_ in dataset:
  mean+= img.mean([1,2])

mean = mean/len(dataset)
print(mean)

#mean = torch.Tensor([0.5082, 0.4643, 0.4186])
#print(mean)
```

```
#finding the standard deviation
sumel = 0.0
countel = 0

for img, _ in dataset:
    img = (img - mean.unsqueeze(1).unsqueeze(1))**2
    sumel += img.sum([1, 2])
    countel += torch.numel(img[0])
  
std = torch.sqrt(sumel/countel)
print(std)
```

```
#libraries
from __future__ import print_function, division

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os
import copy

plt.ion()  
```

```
!pwd
#ensure that you're in the dataset folder
```

```
SPLITS = 5
BATCH_SIZE = 128
```

```
data_transforms = transforms.Compose(
    [
     transforms.Resize((64,64)),
     transforms.RandomHorizontalFlip(),
     transforms.ColorJitter(brightness = 0.25,saturation = 0.25,contrast=  0.25,hue = 0.1),
     transforms.RandomApply([
     transforms.RandomRotation(50),
     ],p=0.25
     ),
     transforms.ToTensor(), #needs to here before the normalization
     transforms.Normalize([0.5082, 0.4643, 0.4186], [0.2871, 0.2846, 0.2957]),
    ]
)
data_dir = './'
dataset = datasets.ImageFolder(os.path.join(data_dir,'train'),transform = data_transforms)
class_names = dataset.classes
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```
#visualizing the training images after transforms
# (key, )
def imshow(inp, title=None):
    """Imshow for Tensor."""

    inp = inp.numpy()
    inp = np.swapaxes(inp,1,3)
    inp = np.swapaxes(inp,1,2)

    mean = np.array([0.5082, 0.4643, 0.4186])
    std = np.array([0.2871, 0.2846, 0.2957])

    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)

    for i in range(len(inp)):
        plt.imshow(inp[i])
        plt.title(title[i])
        plt.show()
    plt.pause(0.001)  # pause a bit so that plots are updated

train_dataloader = torch.utils.data.DataLoader(dataset,batch_size = 10,shuffle = False)
inputs,classes = next(iter(train_dataloader))
imshow(inputs, title=[class_names[x] for x in classes])
```

```
def train_model(model,train_loader,val_loader, criterion, optimizer, scheduler, num_epochs=25,fold = 0):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    patience = 0
    losses = 0
    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)
        
        #for training part
        running_loss = 0.0
        running_corrects = 0.0
        model.train()
        for batch_idx, (inputs,labels) in enumerate(train_loader):
            #print('Train Batch idx : {}, data shape : {}, target shape: {}, target : {}'.format(batch_idx,data.shape,target.shape,target))
            
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            # done before every batch
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                loss = criterion(outputs, labels)
                # backward + optimize only if in training phase
                
                loss.backward()
                optimizer.step()
                # losses +=loss.mean()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)
            
        train_epoch_loss = running_loss / len(train_loader.dataset)
        train_epoch_acc = running_corrects.double() / len(train_loader.dataset)

        print('Training Loss: {:.4f} Training Acc: {:.4f}'.format(
             train_epoch_loss, train_epoch_acc))

        #for validation part
        running_loss = 0.0
        running_corrects = 0.0
        model.eval()
        for batch_idx, (inputs,labels) in enumerate(val_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()

            with torch.set_grad_enabled(False):
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                loss = criterion(outputs, labels)
            # statistics
            
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)
            
        val_epoch_loss = running_loss / len(val_loader.dataset)
        val_epoch_acc =  running_corrects.double() / len(val_loader.dataset)
        scheduler.step()

        print('Validation Loss: {:.4f} Validation Acc: {:.4f}'.format(
             val_epoch_loss, val_epoch_acc))

        # deep copy the model
        if  val_epoch_acc > best_acc:
            print(f"Validation accuracy improved from {best_acc} to {val_epoch_acc}")
            best_acc = val_epoch_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            #save the model weights
            
            print("saving best checkpoint")
            path = './../Resnet50_more_data_aug_'+str(fold)
            torch.save(best_model_wts,path)
            patience = 0
        else:
            patience+=1
            print(f"Validation accuracy didn't improve from {best_acc}")
        print()
        if patience>=10:
            print(f"Patience Expired no improvement seen from validation score of {best_acc}")
            break
    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))
```

```
from sklearn.model_selection import KFold

kf = KFold(n_splits = SPLITS, shuffle = True) #using Kfold for training and testing split

for i,(train_index,test_index) in enumerate(kf.split(dataset)):
  
  #for each fold it generated the set of train_index and test_index
  train = torch.utils.data.Subset(dataset,train_index)
  test = torch.utils.data.Subset(dataset,test_index)

  #dataloader helps to iterate through the dataset, validation set goes throught the same augmentations as the training set.
  train_loader = torch.utils.data.DataLoader(train,batch_size = BATCH_SIZE,shuffle = True, num_workers =0,pin_memory = False)
  val_loader = torch.utils.data.DataLoader(test,batch_size = BATCH_SIZE,shuffle = True, num_workers = 0, pin_memory = False )

  print(f"Fold: {i+1}, Training : {len(train_loader.dataset)}, Validation : {len(val_loader.dataset)}")
  
  #loading the pre-trained model 
  model_ft = models.resnet50(pretrained=True)

  #changing the last layer to have number of classes equal to 4(in our case) from 1000 as trained on imagenet.
  num_ftrs = model_ft.fc.in_features
  model_ft.fc = nn.Linear(num_ftrs, len(class_names))
  model_ft = model_ft.to(device)

  #loss function
  criterion = nn.CrossEntropyLoss()

  # Observe that all parameters are being optimized
  #SGD loss
  optimizer_ft = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
  # Decay LR by a factor of 0.1 every 7 epochs
  exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)
  train_model(model_ft,train_loader, val_loader, criterion, optimizer_ft, exp_lr_scheduler,num_epochs=100,fold = i+1)

  #SGD+Cyclic Learning optimizer
  # optimizer_ft = torch.optim.SGD(model_ft.parameters(), lr=1., momentum=0.9)
  # scheduler_ft = torch.optim.lr_scheduler.CyclicLR(optimizer_ft,step_size_up = 4*len(train_loader), base_lr=3e-3, max_lr=3e-3/6)
  # train_model(model_ft,train_loader, val_loader, criterion, optimizer_ft, scheduler_ft,num_epochs=100,fold = i+1)
```

#**Testing Code**



*   Testing Transforms
*   **predict_image** function
*   All the models trained on different fold are used for predictions



```
test_transforms = transforms.Compose([
  transforms.Resize((80,80)),
  transforms.ToTensor(),
  transforms.Normalize([0.5082, 0.4643, 0.4186], [0.2871, 0.2846, 0.2957]),

])
BATCH_SIZE = 16
```

```

from PIL import ImageOps
def predict_image(model,image):
    """
    converts converts CMYK and grat-scale images(L) to RGB
    pass the input image tensor through the model
    confidence_score and label for each image is generated

    input: model, image in PIL format
    output: class-wise prediction probabilities, class_label
    
    """

    try:
        if image.mode=='CMYK' or image.mode=='L':
            image = image.convert('RGB')
        image_tensor = test_transforms(image).float()
        #print(image_tensor)
    except:
        image = ImageOps.colorize(image,black = "blue",white="white")        
        image_tensor = test_transforms(image).float()
        print('Black n White Image Found.')

    image_tensor = image_tensor.unsqueeze_(0)
    input = torch.autograd.Variable(image_tensor)
    input = input.to(device)
    output = model(input)

    confidence_score = nn.functional.softmax(output,dim = 1)
    label = output.data.cpu().numpy().argmax()

    return label,confidence_score
```

```
#loading all the models and predicting class-wise probabilities for every fold.
import pandas as pd
from PIL import Image
import cv2
#MODEL_NAME = 'Resnet50_more_data_aug_'
MODEL_NAME = 'Resnet_50_best_weights_fold_'
FOLDS =5 

data_dir = './'
test_df = pd.read_csv(os.path.join(data_dir,'test.csv'))
test_dir = os.path.join(data_dir,'Test Images')
for fold in range(1,FOLDS+1):
    fold_image_result = []

    #load the model
    model_ft = models.resnet50(pretrained=False)
    num_ftrs = model_ft.fc.in_features
    model_ft.fc = nn.Linear(num_ftrs, 4)
    model_ft = model_ft.to(device)

    print("Weights File:",MODEL_NAME+str(fold))

    #loading model weights
    weights = torch.load(MODEL_NAME+str(fold),map_location = device)
    model_ft.load_state_dict(weights)

    #putting in evaluation mode
    model_ft.eval()

    for index,row in test_df.iterrows():
      
        img = Image.open(os.path.join(test_dir,row[0]))
        label, confidence_score = predict_image(model_ft,img)
        confidence_score = confidence_score.data.cpu().numpy()[0]
        fold_image_result.append((label,confidence_score))

        if (index)%500==0:
            print(index)
            
    test_df['Fold_'+str(fold)] = pd.Series(fold_image_result)
print(test_df)
```
